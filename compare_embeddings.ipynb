```python
from langchain_openai import OpenAIEmbeddings
from langchain.evaluation import load_evaluator

# Get embedding for a word.
embedding_function = OpenAIEmbeddings()
#embeddings = OpenAIEmbeddings(model='text-embedding-3-small', dimensions=256) 
apple_vector = embedding_function.embed_query("Sumit")
print(f"Vector for 'apple': {apple_vector}")
print(f"Vector length: {len(apple_vector)}")

iphone_vector = embedding_function.embed_query("Venkat")
print(f"Vector for 'iPhone': {iphone_vector}")
print(f"Vector length: {len(iphone_vector)}")

# Compare vector of two words
# This returns a distance score, meaning that the lower the number, 
# the more similar the outputs are, according to their embedded representation.
evaluator = load_evaluator(
    "pairwise_embedding_distance", distance_metric=EmbeddingDistance.COSINE
)
words = ("Sumit", "Venkat")
x = evaluator.evaluate_string_pairs(prediction=words[0], prediction_b=words[1])
print(f"Comparing ({words[0]}, {words[1]}): {x}")
```
